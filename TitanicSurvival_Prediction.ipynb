```python
survived = train_df['Survived']
train_df = train_df.drop(['Survived'], axis=1)
display(train_df)
```

```python
train_index = train_df['PassengerId']
test_index = test_df['PassengerId']

combine_df = pd.concat([train_df, test_df]).reset_index(drop=True)

display(combine_df)
```

```python
print(combine_df.isnull().sum())
```

```python
y = combine_df['Age'].value_counts()
x = y.index.values

plt.figure(figsize = (10,5))
plt.bar (x,y)
plt.title("Ages")
plt.show()

age_class = combine_df [['Pclass', 'Age']]

for i in range (1,4):
  plt.figure(figsize = (10,5))
  plt.title("Class"+str(i))
  y = combine_df[age_class['Pclass']==i]['Age'].value_counts()
  x = y.index.values


  plt.bar (x,y)
  plt.show()

med = age_class.groupby('Pclass').median()
print(med)
```

```python
for i in range(3):
  indx = np.where((combine_df['Pclass']==i+1) & (combine_df['Age'].isnull()))[0]
  combine_df.loc[indx, 'Age'] = med.values[i][0]

  display(combine_df)
```

```python
display(combine_df[combine_df['Fare'].isnull()])
```

```python
same_fare = combine_df [(combine_df['Pclass']==3) & (combine_df['Embarked']=='S')]['Fare']

print('Median', same_fare.median())

plt.hist(same_fare)
plt.show

combine_df['Fare'].fillna(same_fare.median(), inplace=True)
```

```python
display(combine_df[combine_df['Embarked'].isnull()])
```

```python
same_embarked = combine_df [ (combine_df['Pclass'] == 1) & (combine_df['Fare'] >= 70) & (combine_df['Fare'] <= 90) ]['Embarked']

print(same_embarked.value_counts())
combine_df['Embarked'].fillna('C', inplace=True)

display(combine_df)
```

```python
combine_df['Cabin'].fillna('M', inplace= True)
combine_df['Cabin'] = combine_df['Cabin'].str[0]

print(combine_df['Cabin'].value_counts())

indx = np.where(combine_df['Cabin'] == 'T')[0]
combine_df.loc[indx, 'Cabin'] = 'M'
```

```python
from re import T
print(combine_df['Name'].unique())

names = combine_df['Name']

l_name = []
title = []
f_name = []


for name in names:
  if ', ' not in name:
    l_name.append('')
  else:
    last, name = name.split(', ',1 )
    l_name.append(last)


  if '. ' not in name:
    title.append('')
  else:
    titles, first = name.split('. ',1 )
    title.append(titles)
    f_name.append(first)

l_names = np.array(l_name)
titles = np.array(title)
f_names = np.array(f_name)



indx = np.where(np.isin(titles, ['Capt', 'Col', 'Major']))
titles[indx] = 'Military'


indx = np.where(np.isin(titles, ['Don', 'Dona', 'Jonkheer', 'Lady', 'Master', 'the Countess', 'Sir']))
titles[indx] = 'Nobles'

indx = np.where(np.isin(titles, ['Miss', 'Mme', 'Mrs', 'Ms']))
titles[indx] = 'Ms'

indx = np.where(np.isin(titles, ['Miss', 'Mlle' 'Ms']))
titles[indx] = 'Ms'

indx = np.where(np.isin(titles, [ 'Mme', 'Mrs']))
titles[indx] = 'Mrs'

print('\n')

print(np.unique(titles, return_counts=True))

combine_df['title'] = titles
display(combine_df)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
# combine_df.set_index('PassengerId', inplace=True)

display(combine_df)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = combine_df.copy()

label_enc = label_enc.astype(str)

label_enc = label_enc.apply(LabelEncoder().fit_transform)

display(label_enc)
```

```python
one_hot = label_enc.copy()

one_hot = pd.get_dummies(one_hot, columns= ['Sex', 'Embarked', 'title'])

display(one_hot)
```

```python
x = one_hot.loc[train_index].values
y = survived.values

scaler = MinMaxScaler()
scaler.fit(x)
x_scaled = scaler.fit_transform(x)


x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size= .2, random_state= 0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
clf = RandomForestClassifier(random_state=0)

clf.fit(x_train, y_train).score(x_test, y_test)
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'n_estimators': [20, 50, 100, 200, 300, 400, 500, 800, 1000],
    'max_depth': np.arange(3, 50),
    'min_samples_split': np.arange(1, 1000),
    'max_features': ['sqrt', 'log2'],
    'max_samples': np.linspace(0.1, 0.9, 10)
}

# RandomizedSearchCV
rand_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=0),
    params,
    scoring='accuracy',
    random_state=0,
    cv=5
)


rand_search.fit(x_train, y_train)


rand_params = rand_search.best_params_
print("Best Parameters:", rand_params, '\n')


print("Train Acc:", rand_search.best_score_)


preds = rand_search.predict(x_test)
print("Test Acc:", accuracy_score(preds, y_test))
```

```python
n_estimators = np.linspace(rand_params['n_estimators'] - 10, rand_params['n_estimators'] + 10, 3).astype(int)
min_samples_split = np.arange(rand_params['min_samples_split'] - 3, rand_params['min_samples_split'] + 3).astype(int)
max_samples = np.linspace(rand_params['max_samples'] - 0.05, rand_params['max_samples'] + 0.05, 6)
max_depth = np.arange(rand_params['max_depth'] - 5, rand_params['max_depth'] + 5).astype(int)


params = {
    'criterion': [rand_params['criterion']],
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'max_features': [rand_params['max_features']],
    'max_samples': max_samples
}


grid_search = GridSearchCV(
    RandomForestClassifier(random_state=0),
    params,
    scoring='accuracy',
    cv=5
)


grid_search.fit(x_train, y_train)


grid_params = grid_search.best_params_
print(grid_params, '\n')
print("Train Acc:", grid_search.best_score_)


preds = grid_search.predict(x_test)
print("Test Acc:", accuracy_score(preds, y_test))
```
